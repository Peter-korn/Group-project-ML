```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import RobustScaler
```

```python
df = pd.read_excel('Х5.xlsx', engine = 'calamine')
```

```python
df
```

```python
out = pd.DataFrame()
```

```python
out['month_sin'] = np.sin(2 * np.pi * df['Месяц'] / 12)
out['month_cos'] = np.cos(2 * np.pi * df['Месяц'] / 12)
```

```python
rs = RobustScaler()
out[['Численность населения', 'Количество домохозяйств', 'Трафик пеший, в час', 'Трафик авто, в час']] = rs.fit_transform(df[['Численность населения', 'Количество домохозяйств', 'Трафик пеший, в час', 'Трафик авто, в час']])
```

```python
dmms = pd.get_dummies(df[['Дата открытия, категориальный', 'Торговая площадь, категориальный']]).astype(int)
out = pd.concat([out, dmms], axis = 1)
```

```python
out
```

```python
df_cut_average_bill = pd.read_excel('df_cut_average_bill.xlsx')
```

```python
df_cut_traffic = pd.read_excel('df_cut_traffic.xlsx')
```

```python
traffic_final = pd.concat([df_cut_traffic.reset_index(drop = True), out.reset_index(drop = True)], axis = 1)
bill_final = pd.concat([df_cut_average_bill.reset_index(drop = True), out.reset_index(drop = True)], axis = 1)
```
